```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generating hypothetical data
years = np.arange(1960, 2023)
birth_rates = 45 - (years - 1980) ** 2 / 200  # Quadratic trend for demonstration purposes

# Create a DataFrame
data = pd.DataFrame({'Year': years, 'Birth Rate': birth_rates})

# Create a line plot
plt.figure(figsize=(12, 6))
plt.plot(data['Year'], data['Birth Rate'], marker='o', linestyle='-', color='b')

# Customize the plot
plt.title('Hypothetical Birth Rate Trend')
plt.xlabel('Year')
plt.ylabel('Birth Rate (per 1,000 people)')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file_path.csv' with the actual path to your CSV file
file_path = 'India_CSV.csv'

# Read the CSV file into a DataFrame
data = pd.read_csv(file_path)

# Extract relevant columns for plotting
years = data.columns[4:]

# Group the years into decades
year_groups = [f'{decade}-{decade+9}' for decade in range(1960, 2020, 10)]

# Calculate the average birth rate for each decade
average_birth_rates = [data.iloc[:, (decade_idx):(decade_idx+10)].mean().mean() for decade_idx in range(4, len(years), 10)]

# Create a line plot
plt.figure(figsize=(12, 6))
plt.plot(year_groups, average_birth_rates, marker='o', linestyle='-', color='b')

# Customize the plot
plt.title('Average Crude Birth Rate in India (1960-2019)')
plt.xlabel('Decade')
plt.ylabel('Average Birth Rate (per 1,000 people)')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file_path.csv' with the actual path to your CSV file
file_path = 'India_CSV.csv'

# Read the CSV file into a DataFrame
data = pd.read_csv(file_path)

# Extract relevant columns for plotting
years = data.columns[4:]

# Group the years into decades
year_groups = [f'{decade}-{decade+9}' for decade in range(1960, 2020, 10)]

# Calculate the average birth rate for each decade
average_birth_rates = [data.iloc[:, (decade_idx):(decade_idx+10)].mean().mean() for decade_idx in range(4, len(years), 10)]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(average_birth_rates, labels=year_groups, autopct='%1.1f%%', startangle=140)

# Customize the plot
plt.title('Average Crude Birth Rate Distribution in India (1960-2019)')

# Show the plot
plt.show()
```
