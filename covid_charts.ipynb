```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the Excel file
file_path = 'covid.csv'  # Replace with the actual path to your file
data = pd.read_csv(file_path, header=5)  # Use the 6th row as column names

# Print column names to verify
print(data.columns)

# Visualization 1: Cases per million
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Cases per million', data=data, palette='viridis')
plt.title('Cases per Million by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualization 2: Mortality per million
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Mortality per million', data=data, palette='mako')
plt.title('Mortality per Million by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualization 3: Vaccination rates
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Vaccination rates', data=data, palette='rocket')
plt.title('Vaccination Rates by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualization 4: Hesitancy
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Hesitancy', data=data, palette='coolwarm')
plt.title('Hesitancy Rates by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
